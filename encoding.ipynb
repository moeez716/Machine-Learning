```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df['time'].value_counts()
```

```python
# encode the time lable encode with sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le=LabelEncoder()
df['encoded time']=le.fit_transform(df['time'])
```

```python
df
```

```python
df['encoded time'].value_counts()
```

```python
oe=OrdinalEncoder()
df['encoded day']=oe.fit_transform(df[['day']])
df
```

```python
#by catagory
oe=OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded day'] = oe.fit_transform(df[['day']])
df
```

```python
df['encoded day'].value_counts()
```

```python
ohe=OneHotEncoder()
ohe.fit_transform(df[['smoker']]).toarray()
```

```python
# example of one hot encoding
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
onrhot_encoder=OneHotEncoder()
embarked_onehot=onrhot_encoder.fit_transform(titanic[['embarked']])
embarked_onehot_df=pd.DataFrame(embarked_onehot.toarray(), columns=onrhot_encoder.get_feature_names_out(['embarked']))
embarked_onehot_df.head()
titanic=pd.concat([titanic, embarked_onehot_df], axis=1)
titanic.head()
```

```python
#using pandas get_dummies
import pandas as pd
df=sns.load_dataset("tips")
df
```

```python
#pandas get_dummies
pd.get_dummies(df['day'])
#cancate
df=pd.concat([df, pd.get_dummies(df['day'], prefix='day')], axis=1)
```

```python
df
```

```python
#use pandas get dummies it is type of one hot encoding
get_dummies=pd.get_dummies(df,columns=['day'])
get_dummies.head()
```
